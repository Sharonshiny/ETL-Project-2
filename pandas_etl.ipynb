```python
import pandas as pd
from sqlalchemy import create_engine
```

### Extract CSVs into DataFrames

```python
H1B_2015_file = "./Resources/H-1B_FY15.csv"
H1B_2015_df = pd.read_csv(H1B_2015_file, encoding = "ISO-8859-1")
H1B_2015_df.head()
```

```python
H_1B_FY16_file = "./Resources/H-1B_FY16.csv"
H_1B_FY16_df = pd.read_csv(H_1B_FY16_file, encoding = "ISO-8859-1")
H_1B_FY16_df.head()
```

```python
H_1B_FY17_file = "./Resources/H-1B_FY17.csv"
H_1B_FY17_df = pd.read_csv(H_1B_FY17_file, encoding = "ISO-8859-1")
H_1B_FY17_df.head()
```

```python
H_1B_FY17_df.rename(columns={'NAICS_CODE': 'NAIC_CODE'}, inplace=True)
H_1B_FY17_df.head()
```

```python
for col in H_1B_FY17_df.columns: 
    print(col) 
```

```python
H_1B_FY18_file = "./Resources/H-1B_FY18.csv"
H_1B_FY18_df = pd.read_csv(H_1B_FY18_file, encoding = "ISO-8859-1")
H_1B_FY18_df.head()
```

```python
H_1B_FY18_df.rename(columns={'NAICS_CODE': 'NAIC_CODE'}, inplace=True)
```

```python
H1B_2015_selected= H1B_2015_df[['CASE_STATUS','VISA_CLASS','NAIC_CODE']]
H1B_2015_selected.head()
```

```python
H1B_2016_selected= H_1B_FY16_df[['CASE_STATUS','VISA_CLASS','NAIC_CODE']]
H1B_2016_selected.head()
```

```python
H1B_2017_selected= H_1B_FY17_df[['CASE_STATUS','VISA_CLASS','NAIC_CODE']]
H1B_2017_selected.head()
```

```python
H1B_2018_selected= H_1B_FY18_df[['CASE_STATUS','VISA_CLASS','NAIC_CODE']]
H1B_2018_selected.head()
```

```python
Join_H1B = [H1B_2015_selected, H1B_2016_selected, H1B_2017_selected,H1B_2018_selected]
Combined_H1B = pd.concat(Join_H1B)
Combined_H1B
```

```python
Combined_H1B.isnull().sum()
```

```python
Combined_H1B.dropna()
```

```python
Combined_H1B_certAndDenied = Combined_H1B.dropna().query('CASE_STATUS=="CERTIFIED" | CASE_STATUS=="DENIED"')
Combined_H1B_certAndDenied.head()
```

```python
Combined_H1B_certAndDenied_split = pd.get_dummies(Combined_H1B_certAndDenied, columns=['CASE_STATUS'])
Combined_H1B_certAndDenied_split.head()
```

```python
Combined_H1B_agg = Combined_H1B_certAndDenied_split.groupby(["NAIC_CODE"],as_index=False).sum()
Combined_H1B_agg.head()
```

```python
Combined_H1B_agg['PERCENTAGE_DENIED']= (Combined_H1B_agg['CASE_STATUS_DENIED']/(Combined_H1B_agg['CASE_STATUS_CERTIFIED'] + Combined_H1B_agg['CASE_STATUS_DENIED']))
Combined_H1B_agg.head()
```

```python
Finalized_H1B= Combined_H1B_agg
Finalized_H1B
```

```python
us_perm_file = "./Resources/us_perm_visas.csv"
us_perm_df = pd.read_csv(us_perm_file)
us_perm_df.head()
```

```python
us_perm_df.rename(columns={'naics_code': 'NAIC_CODE'}, inplace=True)
```

```python
us_perm_selected= us_perm_df[['case_status','class_of_admission','NAIC_CODE','naics_title']]
us_perm_selected.head()
```

```python
us_perm_selected.isnull().sum()
```

```python
us_perm_selected.dropna()
```

```python
us_perm_certAndDenied = us_perm_selected.dropna().query('case_status=="Certified" | case_status=="Denied"')
us_perm_certAndDenied.head()
```

```python
us_perm_certAndDenied_split = pd.get_dummies(us_perm_certAndDenied, columns=['case_status'])
us_perm_certAndDenied_split.head()
```

```python
us_perm_certAndDenied_agg = us_perm_certAndDenied_split.groupby(["NAIC_CODE"],as_index=False).sum()
us_perm_certAndDenied_agg.head()
```

```python
us_perm_certAndDenied_agg['percentage_Denied']= us_perm_certAndDenied_agg['case_status_Denied']/(us_perm_certAndDenied_agg['case_status_Certified'] + us_perm_certAndDenied_agg['case_status_Denied'])
us_perm_certAndDenied_agg.head()
```

```python
Finalized_US_Perm= us_perm_certAndDenied_agg
Finalized_US_Perm
```
